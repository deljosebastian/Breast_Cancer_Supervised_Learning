## Splitting the dataset

The data is then split into the train and test data sets

```python
from sklearn.model_selection import train_test_split
```

```python
x = df
y = data.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state= 42)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model  = LogisticRegression()
model.fit(X_train,y_train)
```

Model prediction

```python
y_pred  = model.predict(X_test)
y_pred
```

Accuracy score

```python
model1_score = model.score(X_test,y_test)
model1_score
```

# Model Comparison

The comparison of the models can be done by comparing the accuracy scores of the models

```python
results = {
    "logistic Regression" : model1_score,
    "Decision Tree Classifier" : model2_score,
    "Random Forest Classifier" : model3_score,
    "Support Vector Classifier" : model4_score,
    "K Nearest Neighbour" : model5_score
}
```

```python
result_df = pd.DataFrame(list(results.items()), columns=["Model", "Accuracy"])
```

```python
result_df['Accuracy'] = result_df['Accuracy'].round(2)
result_df = result_df.sort_values(by='Accuracy', ascending=False).reset_index(drop=True)
```

```python
result_df
```

## Explanation

As said in the table above, the algorithm with the highest accuracy score is Logistic regression and with the least accuracy score is Support vector classifier
